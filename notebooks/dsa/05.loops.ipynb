we could write each loop as a $ \Sigma $:

* if independant (outer simgma has no effect on the bounds or the inside parameter of the inner sigma):
we could multiply their complexity estimatates to the final answer.

######e.g.

$\text{Do } (i=1, i \leq n, i++) $ 

$\quad \text{Do } (j=1, j \leq n, j=j+i) $ 

$\quad \quad \text{Do } (k=1, k \leq n, k=k*2) $ 

$\quad \quad \quad x++;$ 

$\quad \quad \quad y++;$


seperate the problem into 2 :

* k loop is independant so we would multiply it to the answer of the other parts 

$ k=k*2 \to \sum_{k=1}^n (2) = 2 log_2(n) = \Theta(log(n)) $

* J loop is dependant to the i :$\quad  j=j+i \to \sum_{j=1}^n = \frac{n}{i} $
* I loop : $\quad i++ \to \sum_{i=1}^n \frac{n}{i} = n log(n)$ 

$ \rightarrow \Theta(n \log(n) \cdot \log(n))$
