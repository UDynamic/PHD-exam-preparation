# **Tower of Hanoi: Two Forms with Recurrence Solutions**

## **Form 1: Classic Tower of Hanoi**

### **Problem Statement:**
- 3 pegs: A (source), B (auxiliary), C (destination)
- n disks of different sizes on peg A
- Can only move one disk at a time
- Never place larger disk on smaller disk
- Goal: Move all disks from A to C

---

### **Recurrence Relation:**
Let \(T_1(n)\) = minimum moves for n disks in classic Hanoi

**Reasoning:**
1. Move (n-1) disks from A to B (using C as auxiliary): \(T_1(n-1)\) moves
2. Move largest disk from A to C: 1 move
3. Move (n-1) disks from B to C (using A as auxiliary): \(T_1(n-1)\) moves

\[
T_1(n) = 2T_1(n-1) + 1, \quad T_1(1) = 1
\]

---

### **Solving the Recurrence:**

**Step 1: Homogeneous part:**
\[
T_1(n) - 2T_1(n-1) = 0
\]
Characteristic: \(r - 2 = 0 \Rightarrow r = 2\)
\[
T_1^{(h)}(n) = C \cdot 2^n
\]

**Step 2: Particular solution:**
\(f(n) = 1\) (constant)
Guess: \(T_1^{(p)}(n) = A\) (constant)

Substitute: \(A = 2A + 1 \Rightarrow -A = 1 \Rightarrow A = -1\)
So \(T_1^{(p)}(n) = -1\)

**Step 3: General solution:**
\[
T_1(n) = C \cdot 2^n - 1
\]

**Step 4: Apply initial condition:**
\(T_1(1) = 1\):
\[
C \cdot 2^1 - 1 = 1 \Rightarrow 2C - 1 = 1 \Rightarrow 2C = 2 \Rightarrow C = 1
\]

\[
\boxed{T_1(n) = 2^n - 1}
\]

**Verification:** 
- \(n=1\): \(2^1 - 1 = 1\) ✓
- \(n=2\): \(2^2 - 1 = 3\) ✓ (AB, AC, BC)
- \(n=3\): \(2^3 - 1 = 7\) ✓

---

## **Form 2: Restricted Tower of Hanoi (A→C moves only allowed)**

### **Problem Statement (Reve's Puzzle variant):**
- Same 3 pegs A, B, C
- Same n disks
- **Restriction:** Direct moves from A→C and C→A are **NOT ALLOWED**
- Must always go through B
- Still can't place larger on smaller

This is sometimes called the "Cyclic Hanoi" or "Reve's Puzzle" with restriction.

---

### **Recurrence Relation:**
Let \(T_2(n)\) = minimum moves for n disks with A→C restriction

**To move n disks from A to C:**
1. Move (n-1) disks from A to C (but wait, A→C direct not allowed! Need different strategy...)

Actually, with A→C restriction:
**Strategy:**
1. Move (n-1) disks from A to B (using C): \(T_2(n-1)\) moves
2. Move largest disk from A to B: 1 move (A→B allowed)
3. Move (n-1) disks from B to C (using A): \(T_2(n-1)\) moves
4. Move largest disk from B to C: 1 move (B→C allowed)

But wait, steps 2 and 4 require moving largest disk twice! Actually...

**Correct reasoning:**
To move n disks from A to C with no A→C moves:
1. Move (n-1) disks from A to C: \(T_2(n-1)\) moves
   (But this requires moving them through B since A→C not allowed)
   
Actually, let's think systematically:

Let's define:
- \(T_2(n)\) = moves to transfer n disks from A to C with restriction

**To move n disks from A to C:**
1. Move (n-1) disks from A to B (using C): \(T_2(n-1)\) moves
   (But to move from A to B, that's allowed, no restriction issue)
2. Move largest disk from A to C: **NOT ALLOWED!** So can't do this.

We need a **different approach**. With A→C restriction:

**Actually, here's the correct recurrence:**
We need to move n disks from A to C without direct A→C moves.

One valid strategy:
1. Move (n-1) disks from A to C (but through B somehow): Let's call this \(T_2(n-1)\)
   Actually, to move from A to C without direct moves, we need to go A→B→C
   
Better: Define \(T_2(n)\) = moves to move n disks from A to C with restriction

**Think recursively:**
To move n disks A→C without A→C moves:
1. Move (n-1) disks from A to C: Can't be \(T_2(n-1)\) because that's what we're trying to compute!

Actually, the known recurrence for this restricted version (Reve's Puzzle with A↔C forbidden) is:
\[
T_2(n) = 3T_2(n-1) + 2
\]
With \(T_2(1) = 2\) (A→B→C)

---

### **Why \(T_2(n) = 3T_2(n-1) + 2\)?**

**Reasoning:**
To move n disks from A to C (no A→C):
1. Move (n-1) disks from A to C: \(T_2(n-1)\) moves
   Wait, that doesn't make sense... We need to think differently.

Actually, the known strategy:
1. Move (n-1) disks from A to B (using C): \(T_2(n-1)\) moves
2. Move largest disk from A to C: **NOT ALLOWED!**
   So instead: Move largest disk A→B (1 move)
3. Move (n-1) disks from B to C: \(T_2(n-1)\) moves
4. Move largest disk from B to C: 1 move

That gives: \(T_2(n) = 2T_2(n-1) + 2\)

But actually, step 3 "Move (n-1) disks from B to C" might not be \(T_2(n-1)\) because that's defined for A→C moves.

The correct known recurrence for "no direct moves between source and destination" is actually:
\[
T_2(n) = 3T_2(n-1) + 2
\]
Because:
1. Move (n-1) disks from A to B: \(T_2(n-1)\)
2. Move largest disk A→B: 1
3. Move (n-1) disks from B to C: \(T_2(n-1)\)
4. Move largest disk B→C: 1
5. But wait, we still have (n-1) disks on C that need to go... Actually, this is getting messy.

Let me give you the **actual standard solution** for restricted Hanoi:

---

### **Known Standard Solution:**

For Tower of Hanoi with **A→C moves forbidden** (must go A→B→C):
\[
T_2(n) = 3T_2(n-1) + 2, \quad T_2(1) = 2
\]

**Reason:** To move n disks from A to C:
1. Move (n-1) disks from A to C: \(T_2(n-1)\) moves
   (Yes, this is possible by going A→B→C for each disk recursively)
2. Move largest disk from A to B: 1 move
3. Move (n-1) disks from C to A: \(T_2(n-1)\) moves
4. Move largest disk from B to C: 1 move
5. Move (n-1) disks from A to C: \(T_2(n-1)\) moves

Total: \(3T_2(n-1) + 2\)

---

### **Solving \(T_2(n) = 3T_2(n-1) + 2\):**

**Step 1: Homogeneous:**
\[
T_2(n) - 3T_2(n-1) = 0
\]
Characteristic: \(r - 3 = 0 \Rightarrow r = 3\)
\[
T_2^{(h)}(n) = C \cdot 3^n
\]

**Step 2: Particular:**
\(f(n) = 2\) (constant)
Guess: \(T_2^{(p)}(n) = A\)

Substitute: \(A = 3A + 2 \Rightarrow -2A = 2 \Rightarrow A = -1\)
So \(T_2^{(p)}(n) = -1\)

**Step 3: General solution:**
\[
T_2(n) = C \cdot 3^n - 1
\]

**Step 4: Apply initial:**
\(T_2(1) = 2\):
\[
C \cdot 3^1 - 1 = 2 \Rightarrow 3C - 1 = 2 \Rightarrow 3C = 3 \Rightarrow C = 1
\]

\[
\boxed{T_2(n) = 3^n - 1}
\]

**Verification:**
- \(n=1\): \(3^1 - 1 = 2\) ✓ (A→B, B→C)
- \(n=2\): \(3^2 - 1 = 8\) ✓
- \(n=3\): \(3^3 - 1 = 26\) ✓

---

## **Comparison Summary:**

| Aspect | Classic Hanoi | Restricted Hanoi (No A→C) |
|--------|--------------|--------------------------|
| **Recurrence** | \(T_1(n) = 2T_1(n-1) + 1\) | \(T_2(n) = 3T_2(n-1) + 2\) |
| **Initial** | \(T_1(1) = 1\) | \(T_2(1) = 2\) |
| **Solution** | \(T_1(n) = 2^n - 1\) | \(T_2(n) = 3^n - 1\) |
| **Growth** | Exponential base 2 | Exponential base 3 |
| **For n=3** | 7 moves | 26 moves |
| **For n=5** | 31 moves | 242 moves |

## **Interesting Observations:**

1. **The restriction makes it MUCH harder:** \(3^n - 1\) grows much faster than \(2^n - 1\)
   
2. **Ratio:** \(\frac{T_2(n)}{T_1(n)} \approx \left(\frac{3}{2}\right)^n\) for large n

3. **Both follow the same solution pattern:**
   - Guess particular solution as constant
   - Get homogeneous solution as \(C \cdot r^n\)
   - Combine and use initial condition

4. **The method works perfectly** for both, even though the constants and coefficients differ.

---

## **Why Both Are Solvable with Our Method:**

Both recurrences are of the form:
\[
T(n) = aT(n-1) + b
\]
Where:
- Classic: \(a=2, b=1\)
- Restricted: \(a=3, b=2\)

General solution for \(T(n) = aT(n-1) + b\):
1. Homogeneous: \(T^{(h)}(n) = C \cdot a^n\)
2. Particular: Guess \(A\), get \(A = aA + b \Rightarrow A = \frac{b}{1-a}\) (if \(a \neq 1\))
3. \(T(n) = C a^n + \frac{b}{1-a}\)
4. Use \(T(1)\) to find C

For classic: \(A = \frac{1}{1-2} = -1\), so \(T_1(n) = C \cdot 2^n - 1\)
For restricted: \(A = \frac{2}{1-3} = -1\), so \(T_2(n) = C \cdot 3^n - 1\)

**Same pattern, different constants!**
