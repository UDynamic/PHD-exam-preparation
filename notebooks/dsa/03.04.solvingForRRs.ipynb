General Form of Akra-Bazzi

$ T(n) = a_1 T(b_{1}n) + a_2 T(b_{2}n) + f(n) $

we have to find $P$ such that:

$ \sum a_i b_i^p = 1 \to  a_1 b_1^p + a_2 b_2^p = 1 $

* if $ \sum a_i = 1 \to p=0$
* if $ \sum a_i b_i = 1 \to p=1$
* if for a test p: $ \sum a_i b_i^p < 1 \to \text{lower the} \quad p $
* if for a test p: $ \sum a_i b_i^p > 1 \to \text{increase the} \quad p $

Then:

$ T(n)= \Theta(n^p(1 + \int{\frac{f(x)}{n^{p+1}}dx} )) $ 


* Special case for Akra if $ f(n) = \theta(n^k) $

we try $ p = k \rightarrow \begin{cases} T(n) = \Theta(n^k \log(n)) & \sum a_i b_i^p = 1 \\ T(n) = \Theta(n^k) & \sum a_i b_i^p < 1\end{cases} $

Linear RRs

Standard Form: <br/>$T(n) = a_1T(n-1) + a_2T(n-2) + \cdots + a_kT(n-k) + f(n) $ 

Homogeneous part:<br/> $T(n) = a_1T(n-1) + a_2T(n-2) + \cdots + a_kT(n-k) $ 

none-Homogeneous part (Particular solution):<br/> $f(n) $ 

---
### Step 1: Solve the characteristic equation for Homogeneous Part

$ T(n) - a_1T(n-1) - a_2T(n-2) - \cdots - a_kT(n-k) = 0 $ <br/>
$ T(n) = r^n \rightarrow r^n = a_1 r^{n-1} + a_2 r^{n-2} + \cdots + a_k r^{n-k}$ <br/>
Divide by $r^{n-k} \rightarrow r^k = a_1 r^{k-1} + a_2 r^{k-2} + \cdots + a_k$ <br/>

Rewrite as: <br/> $r^k - a_1 r^{k-1} - a_2 r^{k-2} - \cdots - a_k = 0$
* rewrite the smallest as $r^0$ and then increment the power of r by one for bigger ones

solve for the $r$ : $ \begin{cases}\text{Distinct real roots} & T^{(h)}(n) = C_1 r_1^n + C_2 r_2^n \\ \text{Repeated root r (multiplicity m)} & T^{(h)}(n) = r^n + n r^n\end{cases} $

---
### Step 2: guess Particular Solution
For $f(n) \neq 0 $, guess a form similar to $ f(n) = cn^k \cdot A^n $ in which the $ cn^k $ is complete polynomial with unkown $ c_i $ <br/> 

$ T^{(p)}(n) = (c_1n^k + c_2n^{k-1} + \dots + c_n) A^n$

###e.g. <br/>
for $ f(n)=3n5^n \to cn^k \cdot A^n = 3n^1 \cdot 5^n \to T^{(p)}(n) = (c_1n^1 + c_2) 5^n$ <br/>
for $ f(n)=6n^2 \to cn^k \cdot A^n = 6n^2 \cdot 0^n \to T^{(p)}(n) = (c_1n^2 + c_2n^1 + c_3) 0^n$

Substitute guess into original recurrence to solve for the unknown coefficients.  rewrite $T(n)$ as $T^{(p)}(n) $ and do keep in mind that you want to formulate the equation to look like $T^{(p)}(n)$ after substituiton.

###e.g. <br/>
$ T(n) = 2T(n-1) + 3T(n-2) + 4n$ <br/>
$ T^{(p)}(n) = c_1 n + c_2$ <br/>
substitution: $ c_1 n + c_2 = 2(c_1 (n-1) + c_2) + 3(c_1 (n-2) + c_2) = (5c_1 + 4)n + (-8c_1) + (5c_2)$ <br/>
$ \to  c_1 n + c_2 = (5c_1 + 4)n + (-8c_1) + (5c_2)$

for all n: $ \begin{cases} c_1 = 5c_1 + 4 \\ c_2 = (-8c_1) + (5c_2) \end{cases} \Rightarrow  \begin{cases} c_1 = -1 \\ c_2 = -2 \end{cases} \Rightarrow T^{(p)}(n) = -1 n -2$

## Classic Tower of Hanoi

$T(n) = 2T(n-1) + 1, \quad T(1) = 1$


Step 1: Homogeneous part:<br/>
$T(n) - 2T(n-1) = 0$

Characteristic: <br/>
$r - 2 = 0 \Rightarrow r = 2 \Rightarrow \boxed{T^{(h)}(n) = C \cdot 2^n}$


Step 2: Particular solution:<br/>
$T^{(p)}(n) = A$

Substitute:<br/>
$A = 2A + 1 \Rightarrow -A = 1 \Rightarrow A = -1 \to \boxed{T^{(p)}(n) = -1} $

Step 3: General solution:<br/>
$T(n) = C \cdot 2^n - 1$

Step 4: Apply initial condition:<br/>
$T(1) = 1: C \cdot 2^1 - 1 = 1 \Rightarrow 2C - 1 = 1 \Rightarrow 2C = 2 \Rightarrow \boxed{C = 1}$

Finally:<br/>
$\boxed{T(n) = 2^n - 1}$

## Restricted Tower of Hanoi (No Aâ†’C direct moves)

$T(n) = 3T(n-1) + 2, \quad T(1) = 2$

Step 1: Homogeneous part:<br/>
$T(n) - 3T(n-1) = 0$

Characteristic:<br/>
$r - 3 = 0 \Rightarrow r = 3 \Rightarrow \boxed{T^{(h)}(n) = C \cdot 3^n}$

Step 2: Particular solution:<br/>
$T^{(p)}(n) = A$

Substitute:<br/>
$A = 3A + 2 \Rightarrow -2A = 2 \Rightarrow A = -1 \to \boxed{T^{(p)}(n) = -1}$

Step 3: General solution:<br/>
$T(n) = C \cdot 3^n - 1$

Step 4: Apply initial condition:<br/>
$T(1) = 2: C \cdot 3^1 - 1 = 2 \Rightarrow 3C - 1 = 2 \Rightarrow 3C = 3 \Rightarrow \boxed{C = 1}$

Finally:<br/>
$\boxed{T(n) = 3^n - 1}$
