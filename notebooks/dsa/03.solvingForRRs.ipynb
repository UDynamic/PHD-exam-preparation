$ T(n) = A \cdot T(n-1) + f(n) $

$ A $ : Shape of the Tree (divisions)

$ T(n-1) $ : Height of the Tree

$ f(n) $ : Value of nodes

$ T(n) $ : Sum of all nodes in all levels

---

If the nodes at each level sums to $ f(n) $: 

$ T(n) = L \cdot f(n) $

$ L = 1 + h $


e.g.

$ T(n) = 4 \cdot T(\frac{n}{2}) + n^2 $

$ T(n) = (1 + \log(n))n^2 $

not for:

$ T(n) = 3 \cdot T(\frac{n}{2}) + n^2 $

$ T(n) = n^2 \cdot (1 + \frac{3}{4} + \frac{9}{16} + \dots) = n^2 \cdot (4 \vee c) = \Theta(n^2) $

---

$ T(n) = A \cdot T(n-1) + f(n) $ 

$h = n ,\quad L = 1 + h = n + 1 $

$ T(n) = \sum_{k=1}^n A^{n-k} f(k) $

---
$ T(n) = A \cdot T(n-c) + f(n) $ 

$h = \frac{n}{c} ,\quad L = 1 + h = \frac{n}{c} + 1 $

$ T(n) = \sum_{k=0}^{\frac{n}{c}} A^{n-ck} f(ck) $

---

$ T(n) = A \cdot T(\frac{n}{b}) + f(n) $

$ p = \log_{b}A $

$ T(n)= \Theta(n^p(1 + \int{\frac{f(x)}{n^{p+1}}dx} )) $

e.g.

$ T(n) = 2 \cdot T(\frac{n}{2}) + \frac{n}{\log(n)} $

$ p=\log_{2}2 = 1 $

$ T(n)= \Theta(n^1(1 + \int{\frac{\frac{n}{\log(n)}}{n^{2}}})) $

$ \int{\frac{1}{n\log(n)}} = \log(\log(n))$

$ T(n)= \Theta(n + n \log(\log(n))) = \Theta(n \log(\log(n))) $



---
Integral trick for when it's general form (without $ n^{-1} \vee (n\log(n))^{-1} $): 
* just multiply by $ n $ 

$ \int{n^2 \log(n)} = n^3 \log(n)$

$ \int{\frac{n}{\log(n)}} = \frac{n^2}{\log(n)} $

$ \int{\log(n)} = n\log(n) $

$ \int{\frac{1}{\log(n)}} = n\frac{1}{\log(n)} $

Integral trick for when we have $ n^{-1} $: 
* just multiply by $ \log(n) $

$ \int{\frac{1}{n}} = \log(n) $

$ \int{\frac{1}{n} \log^5(n)} = \log(n) \cdot \log^5(n) = \log^6(n) $

Integral trick for when we have $ (n\log(n))^{-1} $: 
* just multiply by $ \log(\log(n)) $ 

$ \int{\frac{1}{n \log(n)}} = \log(\log(n)) $
