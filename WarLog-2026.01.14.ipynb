In building a **Random Forest**, each tree is constructed using a **Bootstrap sample** from the training data. If `n` is the total number of training samples and `d` is the number of features, on average, how many **unique samples** does each tree use?

3) $ 1 - \left(1 - \left(\frac{1}{n}\right)^d\right) $  
4) $ 1 - \left(1 - \frac{1}{n}\right)^n $

### âœ… **Understanding Bootstrap Sampling:**

- In **Bootstrap sampling**, each tree in the Random Forest is trained on a **sample with replacement** from the full training set of size `n`.
- This means that for each sample in the training set, the **probability of being selected in one Bootstrap sample** is:

  $$
  P(\text{sample is selected}) = 1 - P(\text{sample is not selected in n draws})
  $$

- The probability that a specific sample is **not selected** in one draw is:

  $$
  1 - \frac{1}{n}
  $$

- The probability that it is **not selected in all n draws** is:

  $$
  \left(1 - \frac{1}{n}\right)^n
  $$

- So, the probability that a specific sample **is selected at least once** is:

  $$
  1 - \left(1 - \frac{1}{n}\right)^n
  $$

- Therefore, the **expected number of unique samples** in a Bootstrap sample is:

  $$
  n \cdot \left[1 - \left(1 - \frac{1}{n}\right)^n\right]
  $$

But the question is asking for the **average number of samples used per tree**, which is **approximately 63.2% of n** (as $ 1 - (1 - 1/n)^n \approx 0.632 $ for large n).

$$
\boxed{1 - \left(1 - \frac{1}{n}\right)^n}
$$

Which is **Option 4**.
