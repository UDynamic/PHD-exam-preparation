

## Constraint vs Penalty Equivalence
**Front:** How are the constraint formulation R(θ) ≤ t and penalty formulation L(θ) + λR(θ) related? <br/>
**Back:** They are equivalent via Lagrange multipliers. For every constraint size t, there exists a λ such that solving the constrained problem gives the same solution as minimizing L(θ) + λR(θ). The KKT conditions ensure this duality: at the optimal point, ∇L(θ) = -λ∇R(θ).

## Regularization Path Construction
**Front:** How do you construct a regularization path from the geometric perspective? <br/>
**Back:** Draw constraint regions (diamonds/circles) of increasing sizes t₁ < t₂ < t₃ < ... For each region, find where a loss contour touches it. Plot these contact points as t varies—this trace forms the regularization path, showing how optimal coefficients change with constraint tightness.

## Corner Solutions in L1
**Front:** Why does L1 regularization often produce solutions at diamond corners? <br/>
**Back:** Because diamond corners represent points where one coordinate is exactly zero. When a loss contour touches a diamond at a corner, the solution has sparsity (some coefficients = 0). This occurs because diamond corners "stick out," making them likely contact points for elliptical loss contours.

## Practical Path Finding
**Front:** Given a specific λ, how do you find the corresponding constraint size t geometrically? <br/>
**Back:** 1. Solve min L(θ) + λR(θ) to get θ*. 2. Compute t = R(θ*). This t is exactly the constraint size whose boundary would be touched by a loss contour at θ*. Alternatively, for a given t, the corresponding λ is the Lagrange multiplier at the contact point.
